```python
df = df.drop(columns=['Email Format', 'URL of Web Content', 'Anchor text of URL', 'Location', 'Scam Number', 'Comments', 'Website'])
df
```

```python
existing_df = pd.read_csv('final_data.csv')

# Append the new DataFrame to the existing DataFrame
appended_df = pd.concat([existing_df, df], ignore_index=True)

# Save the updated DataFrame to the same Excel file
appended_df.to_csv('final_data.csv', index=False)

print("Data appended and Excel file updated.")
```
