```python
import numpy as np
import pandas as pd
import email
```

```python
df = pd.read_excel("enron_data.xlsx")
```

```python
df
```

```python
print(df.loc[1]['message'])
```

```python
def get_field(field, messages):
    column = []
    for message in messages:
        e = email.message_from_string(message)
        column.append(e.get(field))
    return column

def body(messages):
    column = []
    for message in messages:
        e = email.message_from_string(message)
        column.append(e.get_payload())
    return column
```

```python
df['subject'] = get_field("Subject", df['message'])
df['sender'] = get_field("X-From", df['message'])
df['sender_email'] = get_field("From", df['message'])
df['body'] = body(df['message'])
```

```python
df = df.drop(columns=['file', 'message'])
```
